```python
metadata_903 = MetaData()
 
episodes = Table('episodes', metadata_903, autoload_with=engine_903)
 
print(episodes.name)
print(episodes.c.keys())
print(repr(episodes))
 
```

```python
metadata_chinook = MetaData()
 
albums = Table('albums', metadata_chinook, autoload_with=engine_chinook)
 
print(albums.name)
print(albums.c.keys())
print(repr(albums))
 
```

```python
import pandas as pd

header = Table('header', metadata_903, autoload_with=engine_903)

# REFU NOBT

with engine_903.connect() as con:
    stmt = select(header).where(or_(header.columns.ETHNIC == "REFU", 
                                    header.columns.ETHNIC == "NOBT"))

    result_proxy = con.execute(stmt)
    result = result_proxy.fetchmany(size=20)

df_903 = pd.DataFrame(result)
result
```
